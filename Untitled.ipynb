```python
def calc_dice(X, Y):
    if X.any() or Y.any():
        d = 2* np.count_nonzero(np.logical_and(X, Y)) / (np.count_nonzero(X)+np.count_nonzero(Y))
    else:
        d=0
    return d

def main(idx0):
    print(idx0)
    gt_file = 'D:\BiGART_inputdata\OxyTarget_both_delineations_new.h5'
    
    src = 'D:\BiGART_results'
    files = [os.path.join(src, r'bigart_{}'.format(idx0+k), 'merge_images.h5') for k in range(5)] \
                + [os.path.join(src, 'voted_{}.h5'.format(idx0))]
    
    gt = h5py.File(gt_file, 'r')
    
    p_include = [i for i in gt['y_1'].keys()]
    
    res = []
    for file_path in files:
        with h5py.File(file_path, 'r') as f:
            
            for patient in f['y'].keys():
                print('\t', patient)
                if patient in p_include:
                    prediction = f['predicted'][patient][:,:,:,0]>0.5

                    y_g = gt['y_1'][patient][:,:,:] >0
                    y_g2 = gt['y_2'][patient][:,:,:] >0

                    res.append((patient, 
                                calc_dice(prediction, y_g),
                                calc_dice(prediction, y_g2),
                                calc_dice(y_g, y_g2)
                               ))
    gt.close()
    
    pd.DataFrame(res, columns=['patient_idx', 'P_G1', 'P_G2', 'G1_G2'] ).to_csv('GT_variation_{}.csv'.format(idx0))
    
main(55)
```
